# image check with the box point

```
import matplotlib.patches as patches
import cv2
from google.colab.patches import cv2_imshow
```

```
original_img = cv2.imread('/content/holo.png')
cv2_imshow(original_img)
```

```
box_1
```

```
rectangle_box = cv2.rectangle(original_img, (box_1[0],box_1[1]), (box_1[2], box_1[3]), (255,255,0), 2)
```

```
cv2_imshow(rectangle_box)
```

# create blured image

```
# img[y:y+h, x:x+w]
blur_target = original_img[box_1[1]:box_1[3], box_1[0]:box_1[2]]
```

```
blur_img = cv2.GaussianBlur(blur_target, (15,15), 5)
```

```
cv2_imshow(blur_img)
```

# merge the blur img to original img

```
original_img = cv2.imread('/content/holo.png')

original_img[box_1[1]:box_1[3], box_1[0]:box_1[2]] = blur_img
cv2_imshow(original_img)
```
